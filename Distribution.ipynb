**Q1.** Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use BINOM.DIST)

**Ans-**  

- **Step-1:** **Simulate 30 Dice rolls.**
   
   In **column A**, generate 30 rolls,

   In cell **A1**, type:
    
  **=RANDBETWEEN(1,6)**

  Drag this formula down to A30.

  This will generate random numbers from 1 to 6.

- **Step-2:** **Count How many times 3 appears.**
  
  In another cell **C2**, type:

  **=COUNTIF(A1:A30, 3)**

  This tell you how many times **3** occured in the 30 rolls.

- **Step-3:** **Find Probability of Getting Exactly 5 Threes.**

  Use the **BINOM.DIST** formula.

  In cell **C5**, type:

  **=BINOM.DIST(5, 30, 1/6, FALSE)**

  **Explanation:**

- **5** → exactly 5 times rolling a 3
- **30** → total rolls
- **1/6** → probability of rolling a 3 on one roll
- **FALSE** → exact probability (not cumulative)

  **Result:** 0.19210813 (approximately)

  So, the probability of rolling a 3 exactly 5 times in 30 rolls = **19.21%**

  Z = 55 - 50 / 5

  **Z = 1**

     **Z = -1** and **Z = 1**

  ## **Step-2:** **Use Empirical Rule:**

  In a normal distribution:

- About **68%** of values lie within **1 standard deviation** from the mean.

  Since **45 to 55** = **50 ± 5** (one standard deviation)

**Q4.**  What is the concept of standardization (z-score), and why is it important in data analysis? Explain the formula and how standardization transforms a dataset.

**Ans-** **STANDARDIZATION (Z-SCORE):** It is a technique which is used to scale numerical data into a common format so that different variables can be compared easily.

A **z-score** tells us how far a value is from the mean in terms of standard deviations.

   **Interpretation of Z-score:**

- **Positive z-score (+)** → value is **above the mean**
- **Negative z-score (−)** → value is **below the mean**
- **z = 0** → value is **equal to the mean**

 **After Standardization:**

- **Mean** = 0
- **Standard Deviation** = 1

  In it camparison is only possible when range is equal.

  ## **Why it is important in data analysis?**

  **1. Helps compare different dataset.**
    
    Different datasets may have different units.

   Example:

- Exam marks out of 100
- Salary in rupees

  Z-score converts both to the same scale, making comparison possible.

  **2.Helps detect outliers.**

   Very high or very low z-scores may indicate unusual values (outliers).

  **3. Important in Machine Learning.**

   Algorithms work better when variables are on similar scales.

   Example:

- Age = 25
- Salary = 50,000

  Without standardization, salary dominates the model.

  **4. Used in Normal Distribution.**

  Z-score helps find probabilities using the standard normal table (Z-table).

  ## **FORMULA OF STANDARDIZATION (Z-SCORE)**

  The Z-Score formula is used to standardized data.

  **Z = X - Mu / Sigma**

  Where,

  **Z** = Standardized value

  **X** = Actual / Original value

  **Mu** = Mean (Average) of dataset

  **Sigma** = Standard Deviation

  Example:

   - **Mean** = 50

   - **Standard Deviation** = 5

   - **Value** = 40

    **Apply Formula**

    Z = X - Mu / Sigma

    Z = 40 - 50 / 5

    Z = -10 / 5

    **Z = -2**

    The value of **40 is -2 Standard deviations below the Mean**.

    Because the z-score is **negative**, the value lies **below the average**.

    ## **How standardization transforms a dataset**

    Standardization transforms a dataset by converting original values into Z-Score.

    Example:

    | **Original Values** |
    |--------------------:|
    |         40          |
    |         50          |
    |         60          |
    |         70          |

    **Mean** = 55
    
    **Standard Deviation** = 11.18034

   **Step-1:** **Find Standardized values**

  - **For 40:**
      
    Z = X - Mu / Sigma = 40 - 55 / 11.18034 = **-1.34164**

  - **For 50:**

    Z = X - Mu / Sigma = 50 - 55 / 11.18034 = **-0.44721**

  - **For 60:**
   
    Z = X - Mu / Sigma = 60 - 55 / 11.18034 = **0.447214**

  - **For 70**

    Z = X - Mu / Sigma = 70 - 55 / 11.18034 = **1.341641**

    **Dataset After Standardization**

    | Original Value | Standardized Value (Z-score) |

**Q5.** What is Kurtosis and their type?

**Ans-** Kurtosis is a measure of **“tailedness”** in a distribution. It tells us whether the distribution has a **high peak**, **flat peak**, or **normal peak** and **how heavy the tails are**.

**Types of Kurtosis:**

There are 3 types of Kurtosis-

| Type        | Peak       | Tails    | Kurtosis Value |
| ----------- | ---------- | -------- | -------------: |
| Mesokurtic  | Normal     | Moderate |            = 3 |
| Leptokurtic | High/Sharp | Heavy    |            > 3 |
| Platykurtic | Flat       | Light    |            < 3 |


**1.** **Mesokurtic**

- The distribution has a **normal peak**.
- **Moderate tails**.
- Similar to a normal distribution.
- Kurtosis value = 3

**Shape:** Normal bell-shaped curve.

**2.** **Leptokurtic**
- The distribution has a **high/sharp peak**.
- **Heavy tails**.
- More extreme values (Outlier) present.
- Kurtosis value greater than 3

**Shape:** Tall and narrow peak.

**3.** **Platykurtic**
- The distribution has a **flat peak**.
- **Light tails**.
- Few outliers.
- Kurtosis value less than 3

**Shape:** Flat and wide curve.

**Q6.** Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

**Ans-** A uniform distribution is a distribution in which **all outcomes have equal probability**.

  A fair die has 6 faces:

   **1,2,3,4,5,6**

   Each number has an **equal chance of occurring**.

   Probability of each outcome:

   P(1)=P(2)=P(3)=P(4)=P(5)=P(6)= **1/6**

Since every outcome is **equally likely**, the die follows a **uniform distribution**.

  ## **Why it is a Good Model?**
  
  **1.** **Equal Probability**
  
   Every face of a fair die has the same probability: **1/6**

  **2.** **No Bias**

  A fair die is unbiased, meaning no number is more likely than another.

  **3.** **All Outcomes are Equally Distributed**

  Over many rolls, each number appears approximately the same number of times.

  **Example:** If a die is rolled 600 times, each number is expected to appear about:

  = 600 / 6

  =100 times

  **Uniform distribution is a good model for rolling a fair die because all 6 outcomes (1–6) are equally likely, and each outcome has the same probability of 1/6.**

**Q8.** How does log transformation help in stabilizing variance and making data more normally distributed?

**Ans-** **LOG TRANSFORMATION:** It is a data transformation technique which is used to replace each value X with its logarithm.

   **Y = log(X)**

   Where,

   **Y** = Transformed value(log of X)

   **X** = Original value

   ## **How it stabilizes variance**

   **Example:**

   | **Original Salaries** |
   |----------------------:|
   |        20,000         |
   |        25,000         |
   |        30,000         |
   |        40,000         |
   |        1,00,000       |

   After applying **log**:

   | **Original Salaries** | **Log(Salaries)** |
   |----------------------:|------------------:|
   |     20,000            |     4.30103       |
   |     25,000            |     4.39794       |
   |     30,000            |     4.477121      |
   |     40,000            |     4.60206       |
   |     1,00,000          |     5             |

   Large values are compressed more than small values.

   **Result:** The variability becomes more consistent across the dataset, which is called **variance stabilization**.

   ## **Makes Data More Normally Distributed**

  Many real-world datasets such as income, sales, and population are **right-skewed**.

- Most observations are small or moderate.
- A few observations are extremely large.

  Log transformation reduces the effect of these very large values and pulls the distribution toward symmetry.

  **Result:** The distribution becomes closer to a **normal (bell-shaped) distribution**.
